```python
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plot
```

```python
boston = load_boston()
boston_panda = pd.DataFrame(boston.data)
print(boston_panda.head())
```

```python
def draw_hist(col):
    to_draw = boston_panda.loc[:, col]
    to_draw.plot.hist(bins=20)
```

```python
draw_hist(8)
draw_hist(5)
```

```python
def draw_plot(index):
    x = boston.data[:, index].reshape(-1, 1)
    
    regr = LinearRegression().fit(x, boston.target)
    
    pred = regr.predict(x)
    
    plot.scatter(x, boston.target)
    plot.plot(x, pred, color='y')    
```

```python
draw_plot(5)
```

```python
draw_plot(8)
```
